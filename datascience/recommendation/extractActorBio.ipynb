```python
path_to_json = './data/person/themoviedb/'
json_files = [path_to_json+pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
len(json_files)

theMovieDBPersons=[]

for fname in json_files:
    with open(fname, 'r') as f:
        r = json.load(f)

        extract ={'_actorID': r['_actorID'], 'actor': r['name'], 'gender': r['gender'], 'birthday': r['birthday'],'biography': r['biography'] }
        theMovieDBPersons.append(extract)
        
```

```python
path_to_json = './data/person/imdb/'
json_files = [path_to_json+pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
len(json_files)

IMDBPersons=[]

for fname in json_files:
    with open(fname, 'r') as f:
        r = json.load(f)   
        columns= {'_actorID': '_actorID', 'mini biography': 'biography', 'height':'height', 'salary history': 'salaryhistory' }
        
        extract ={}
        for key in columns:
            try: 
                extract.update({ columns[key] : r[key] } )
            except Exception as e:
                extract.update({ columns[key]  : '' } )
                logger.info('Unable to find key : '+ key + '::' +str(e))

        IMDBPersons.append(extract)
```

```python
path_to_json = './data/person/wikipedia/'
json_files = [path_to_json+pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
len(json_files)

WikiPersons=[]

for fname in json_files:
    with open(fname, 'r') as f:
        r = json.load(f)   
        columns= {'_id': '_actorID', 'Career': 'career' }
        
        extract ={}
        for key in columns:
            try: 
                extract.update({ columns[key] : r[key] } )
            except Exception as e:
                extract.update({ columns[key]  : '' } )
                logger.info('Unable to find key : '+ key + '::' +str(e))

        WikiPersons.append(extract)
```

```python
theMovieDF = pd.DataFrame.from_dict(theMovieDBPersons , dtype=str)
IMDBDF = pd.DataFrame.from_dict(IMDBPersons , dtype=str)
wikiDF = pd.DataFrame.from_dict(WikiPersons , dtype=str)
```

```python
from datetime import date 
  
def calculateAge(birthDate): 
    today = date.today() 
    age = today.year - birthDate.year - ((today.month, today.day) < (birthDate.month, birthDate.day)) 
  
    return age 

celebrity =    theMovieDF.merge( IMDBDF, how='inner', left_on = '_actorID', right_on = '_actorID')
celebrity = celebrity[['_actorID','actor', 'birthday', 'gender', 'height']]
```

```python
a= celebrity.head()
```

```python
celebrityBio=[]

for index, row in celebrity.iterrows():
    _actorID=row['_actorID']
    actor=row['actor']
    birthdate = datetime.strptime(row['birthday'], '%Y-%m-%d')
    bdate= row['birthday']
    age= calculateAge(birthdate)
    gender= row['gender']
    if gender =='1':
        sex='female'
    elif gender=='2': 
        sex='male'
    else:
        sex='na'
    height = row['height']
    celebrityBio.append({'_actorID': _actorID, 'actor': actor, 'birthdate': bdate, 'age':age, 'gender': sex, 'height': height})

f=open('celebrityBio.json', 'w')
f.write(json.dumps(celebrityBio))
f.close()
```
